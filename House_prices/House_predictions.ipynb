```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```
df = pd.read_csv('House_prediction.csv')
```

```
df
```

```
def floor_converter(x):
  if x == '-':
    return 0;
  else:
    return int(x)

def if_animal(x):
  if x == 'accept':
    return 1;
  else:
    return 0;

def if_furnished(x):
  if x == 'furnished':
    return 1;
  else:
    return 0;
```

```
floor_int = lambda x: floor_converter(x) 
animal_int = lambda x: if_animal(x)
furniture_int = lambda x: if_furnished(x)

df.floor = df.floor.apply(floor_int)
df.animal = df.animal.apply(animal_int)
df.furniture = df.furniture.apply(furniture_int)

df.rename(columns={'parking spaces':'parking_spaces',
                   'hoa (R$)':'hoa',
                   'rent amount (R$)':'rent',
                   'property tax (R$)':'property_tax',
                   'fire insurance (R$)':'fire_insurance',
                   'total (R$)':'total'
                  }, inplace='true')

df
```

```
from sklearn.preprocessing import OneHotEncoder 
from sklearn.compose import ColumnTransformer
```

```
def tokenize(temp):

  word_index = {}
  city_int = []
  j = 0

  for i in temp:
    if i in word_index:
      city_int.append(word_index[i])
    else:
      word_index[i] = j
      city_int.append(word_index[i])
      j=j+1

  return word_index, city_int
```

```
temp = np.array(df.city)
word_index, city_int = tokenize(temp)
print(word_index)
df.city = city_int
df
```

```
df_onehot = pd.get_dummies(df.city, prefix = 'city')
print(df_onehot.tail(10))
```

```
df_final = df_onehot.join(df)
df_final
```

```
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
```

```
X = df_final[['fire_insurance']]
y = df_final[['rent']]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```
regressor.score(X_test, y_test)
```

```
y_pred = regressor.predict(X_test)
y1= np.array(y_test).flatten()
y2= np.array(y_pred).flatten()
df_pred = pd.DataFrame({'Actual': y1, 'Predicted': y2})
df1 = df_pred.head(25)
df1.plot(kind='bar',figsize=(16,10))
plt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.show()
```
