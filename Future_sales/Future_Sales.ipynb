```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```
df = pd.read_csv('sales_train.csv')
df_items = pd.read_csv('items.csv')
df_shops = pd.read_csv('shops.csv')
df_test = pd.read_csv('test.csv')
```

```
df = df.drop(columns=['date'])
```

```
df = df.drop(columns=['item_price'])
```

```
temp = df.groupby(by = ['date_block_num', 'shop_id', 'item_id'])
df = temp.sum().reset_index()
df.rename(columns={'item_cnt_day':'item_cnt_month'}, inplace=True)
```

```
df.date_block_num = df.date_block_num.apply(lambda x: x%12)
df
```

```
df_items
```

```
j = 0
for i in df.item_id:
  df.item_id[j] = df_items.loc[df_items['item_id'] == i, 'item_category_id']
  j = j + 1
  if j%100000 == 0:
    print(j,' item_ids processed')

df
```

```
temp_onehot = pd.get_dummies(df.shop_id, prefix = 'shop')
temp_onehot = temp_onehot[0:2000000]
a = df[0:2000000]
a = a.join(temp_onehot)
a
```

```
temp_onehot = pd.get_dummies(df.item_id, prefix = 'item')
temp_onehot = temp_onehot[0:2000000]

a = a.join(temp_onehot)
a
```

```
a = a.drop(columns=['shop_id', 'item_id', 'date_block_num'])
```

```
a
```

```
y = a.item_cnt_month
a = a.drop(columns=['item_cnt_month'])
X = a
```

```
X = np.array(X)
y = np.array(y)
print(X.shape,'\n',y.shape)
```

```
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)
```

```
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```
regressor.score(X_test, y_test)
```
